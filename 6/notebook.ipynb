```python
def update_fishes(fishes, n_days):
    
    fishes_at_day = fishes
    
    for day in range(n_days):
        new_fishes = 0
        updated_list = []
        for i, fish in enumerate(fishes_at_day):
            match fish:
                case 0:
                    updated_list.append(6)
                    new_fishes +=1
                case _:
                    updated_list.append(fish-1)
                    
        fishes_at_day = updated_list + [8]*new_fishes
    return len(fishes_at_day)
```

```python
%time update_fishes(fishes, 80)
```

```python
def update(fishes, n_days):
    if n_days > 0:
        new_fishes = [(x-1)%6 for x in fishes]
        new_fishes.extend([8]*new_fishes.count(6))
        print(new_fishes)
        return update(new_fishes, n_days -1)
    return len(fishes)

%time update(fishes, 80)
```
